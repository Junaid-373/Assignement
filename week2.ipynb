## Q.8 - Discuss the importance of dictionary keys being immutable and provide examples.


*   if mutable types (like lists or dictionaries) were allowed as keys, it could lead to confusing bugs and unexpected behavior. Immutability helps prevent these situations by enforcing clear rules about what can be used as a key.
*   Tuples, which are immutable, can be used as keys. This allows for the creation of composite keys, where multiple values can be combined into a single key.
*   Immutability helps maintain consistency in how data structures behave. If all keys are immutable, developers can have a clear understanding of how data can be manipulated and accessed without side effects.
EXAMPLE --


```
location_map = {
    (10, 20): "Location A",
    (30, 40): "Location B",
    (50, 60): "Location C"
}

# Accessing values using tuple keys
print(location_map[(10, 20)])  # Output: Location A
print(location_map[(30, 40)])  # Output: Location B
```
